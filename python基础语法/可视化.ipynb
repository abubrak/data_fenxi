```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams ['font.sans-serif']= ['SimHei'] # 用来设置字体样式以正常显示中文标签
plt.rcParams ['axes.unicode_minus']=False # 默认是使用Unicode负号，设置正常显示字符，如正常显示负号
```

```python
x=np.arange(0,600)
y=np.sin(x*0.1)+np.random.random(600)
y2=np.sin(x*0.1)+np.random.random(600)
```

```python
fig1=plt.figure(figsize=(4,2),dpi=200)
plt.plot(x,y,linestyle=':',linewidth=1,label='$V_T$')
plt.plot(x,y2,linestyle='-',linewidth=1,label='$V_T$2')
plt.xlim(0,600)
plt.legend(loc=1)
plt.xlabel('Time(s)')
plt.ylabel('$V_T$(V)')
plt.show()
```

# 其他图形绘制

## 散点图

```python
X=np.arange(100)
y=np.random.randint(0,20,100)
plt.scatter(X,y,marker='*',
           s=y*100,
           color='#616161',
           alpha=0.1) 
plt.show()
```

```python
X=np.arange(100)
y=np.random.randint(0,20,100)
plt.scatter(X,y,marker='*',
           s=y*100,
           color='#616161',
           alpha=y/20) #根据y值设置大小
plt.show()
```

## 柱状图

```python
X=np.arange(10)
y=np.random.randint(0,20,10)
plt.bar(X,
        y,
        width=0.2,
        alpha=0.5)
plt.show()
```

## 簇型柱状图

```python
X=np.arange(10)
y=np.random.randint(0,20,10)
y1=np.random.randint(0,20,10)
plt.bar(X,
        y1,
        width=0.2,
        alpha=0.5)
plt.bar(X,
        y,
        width=0.2,
        alpha=0.5)
plt.bar(X,
        y1,
        width=0.2,
        alpha=0.5)
plt.show()
```

```python
X=np.arange(10)
y=np.random.randint(0,20,10)
y1=np.random.randint(0,20,10)
plt.bar(X-0.2,#移动位置交错开
        y1,
        width=0.2,
        alpha=0.5)
plt.bar(X,
        y,
        width=0.2,
        alpha=0.5)
plt.bar(X+0.2,
        y1,
        width=0.2,
        alpha=0.5)
plt.show()
```

## 堆积柱状图

```python
X=np.arange(10)
y=np.random.randint(0,20,10)
y1=np.random.randint(5,20,10)
plt.bar(X,y,width=0.8,color='#1976d2',label='男生')
plt.bar(X,y,width=0.8,color='#008800',label='女生')
plt.xticks(X,[f'{i}年' for i in range(2015,2025)])
plt.legend()
plt.show()
```

## 扇形图

## 箱线图

```python
x=np.random.random(100)
x[0]+=1
x2=np.random.random(100)
plt.boxplot([x,x2])
plt.xticks([1,2],['F','M'])
plt.show()
```

# 3D绘图

# 子图 

```python
fig=plt.figure()
for i in range(1,7):
    fig.add_subplot(2,#行数
                    3,#列数
                    i)#第几个位置
```

```python
fig=plt.figure()
fig.add_subplot(2,2,1)
fig.add_subplot(2,2,2)
fig.add_subplot(2,2,(3,4))
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(2,2,1)
y=np.random.random(3)
ax.pie(y,
        autopct='%.2f',
        pctdistance=0.5,
        explode=[0.01,0.01,0.1],
        colors=list('rgb'))
ax1=fig.add_subplot(2,2,3)
x=np.arange(0,600)
y=x
ax1.plot(x,y)
ax2=fig.add_subplot(2,2,(2,4))
x=np.arange(4)
y=np.random.randint(10,20,4)
ax2.bar(x,
        y)
plt.show()
```

```python
fig=plt.figure()
fig.add_subplot(4,4,1)
plt.pie([1,2,3])
plt.title('子图')
plt.xlabel('子图实例')
ax1=fig.add_subplot(2,2,3)
x=np.arange(0,600)
y=x
ax1.plot(x,y)
ax2=fig.add_subplot(1,2,2)
x=np.arange(4)
y=np.random.randint(10,20,4)
ax2.bar(x,y)
plt.suptitle('总标题')
plt.show()
```

# pandas绘图

```python
data=pd.read_excel('Excel数据.xlsx')
data
```

```python
sale_data=data.groupby('地区')['销售额'].sum()
```

```python
sale_data
```

```python
#plt.bar(sale_data[0],sale_data[1]) 错误示范
```

```python
plt.bar(sale_data.index,sale_data.values)
plt.xlabel('地区')
plt.ylabel('销售额')
plt.show()
```

```python
sale_data.plot.bar()
plt.ylabel('销售额')
plt.title('不同地区的销售总额')
plt.grid()
plt.text(4,sale_data.max(),'销冠')
plt.show()
```

```python
sale_data.plot.bar(title='不同地区的销售总额',
                    ylabel='销售额',
                   grid='True'
)
plt.show()
```
