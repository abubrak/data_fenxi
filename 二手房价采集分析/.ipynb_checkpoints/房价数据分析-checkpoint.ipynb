```python
## 提取数据
for i in ['面积','建房时间','单价','总价','楼层位置']:
    data[i]=data[i].apply(lambda x:float(re.findall(r'-?\d+\.\d+|-?\d+',x)[0]))
```

```python
data.head()
```

## 处理户型

```python
def fun(x):
    try:
        rooms=re.findall(r'\d',x)
        return (rooms[0],rooms[1])
    except:
        return 0,0

data['室'],data['厅']=zip(*data['户型'].apply(fun))#将结果转化为独立的两列
```

```python
data.head()
```

### 房屋标题可视化

```python
#pip install jieba
```

```python
import jieba
```

```python
data['分割标题']=data['房源标题'].apply(lambda x:jieba.lcut(x))
```

```python
ans_dict={}    #存储每个分词单价总和
ans_num_dict={}#存储每个分词出现次数
for i in data.index:
    for j in data.loc[i,'分割标题']:
        if j in ans_dict.keys():
            ans_dict[j]+=data.loc[i,'单价']
            ans_num_dict[j]+=1
        else:
            ans_dict[j]=data.loc[i,'单价']
            ans_num_dict[j]=1

price_dict={key: ans_dict[key]/ans_num_dict[key] for key in ans_dict if key in ans_num_dict and ans_num_dict[key]!=0}
```

```python
price_dict
```

```python
from wordcloud import WordCloud
from PIL import Image
```

```python
# 创建词云对象，设置掩模
wordcloud=WordCloud(font_path=r'C:\Windows\Fonts\simhei.ttf',
                   mask=mask_image,
                   background_color='white').generate_from_frequencies(price_dict)

plt.axis('off')#不显示x y轴
ax=plt.imshow(wordcloud)
fig=ax.figure
fig.set_size_inches(12,12)#可调节图片紧密，尺寸程度
plt.show()
```

### 各地价钱排名前三的小区

```python
def apart_fun(x):
    ans_=x.groupby('小区')['单价'].mean()
    ans_.sort_values(ascending=False,inplace=True)
    return ans_[:3]

apartment_top3=data.groupby('城市').apply(apart_fun)
```

```python
apartment_top3
```

```python
ax=plt.subplot(111,projection='polar')
r=np.linspace(0,2*np.pi,24,endpoint=False)
plt.bar(r,apartment_top3,width=0.2,color=[i for i in color_list for j in range(3)])
#禁用网格线
ax.grid(False)
#禁用圆形边框
ax.set_frame_on(False)
#隐藏半径刻度
ax.set_rgrids([],[],color='white')
#禁用角度标签
ax.set_thetagrids([],[])
#显示特定角度标签
for i in range(24):
    plt.text(r[i],
            apartment_top3.max(),#标签的径向量，使得数据的最大值确保标签显示在图标边缘
             apartment_top3.index.get_level_values(1).tolist()[i],
             rotation=r[i]*180/np.pi,
             ha='center',va='center'
            )

plt.show()
```

## 房屋价格预测

### 映射字典处理

```python
# 城市处理
city_ans.sort_values(inplace=True)
city_dict={item:i for i,item in enumerate(city_ans.index)}
city_dict
```

```python
#方向处理
direction_data.sort_values(inplace=True)
direction_dict={item:i for i,item in enumerate(direction_data.index)}
```

```python
#城市级别处理
level_dict={'准一线':0,'一线':1}
```

### 数据映射

```python
data['city']=data['城市'].map(city_dict)
data['dir']=data['朝向'].map(direction_dict)
data['level']=data['城市等级'].map(level_dict)
data.head()
```

## 提取数据

```python
X,Y=data[['city','dir','level','室','厅','面积','楼层位置','建房时间']],data['单价']
X=X.astype(float)
```

##  模型预测

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#留出法划分数据集
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.7,random_state=42)#设定一个随机种子
```

```python
# 线性回归模型实例
model=LinearRegression()
#训练模型
model.fit(train_x,train_y)
#在测试集测试
y_pre=model.predict(test_x)
#计算预测均方误差
mse=mean_squared_error(test_y,y_pre)
print('均方误差:',mse)
```

```python
# 打印模型参数
print(f'k: {model.coef_}')
```

```python
# 打印截距
print(f'b: {model.intercept_}')
```

```python
#plt.scatter(test_y,y_pre) 线性图
plt.scatter(test_y,y_pre,alpha=0.1)#散点图
plt.show()
```

```python
#pip install xgboost
```

```python
import xgboost as xgb
# 创建XGBoost模型实例
# 损失函数，迭代次数，随机种子
xgb_reg=xgb.XGBRegressor(objective='reg:squarederror',n_estimators=100,seed=42)
# 训练模型
xgb_reg.fit(train_x,train_y)
# 在测试集上进行预测
y_pred=xgb_reg.predict(test_x)
# 计算预测的均方误差
mse=mean_squared_error(test_y,y_pred)
print('均方误差：',mse)
```
