```python
from sklearn.datasets import load_iris
sklearn.__version__
```

```python
iris=load_iris()
X=iris['data']
y=iris['target']
X.shape
```

### 聚类

```python
from sklearn.decomposition import PCA
X_new=PCA(n_components=2).fit_transform(X)
```

```python
X_new.shape
```

### 标准化

```python
from sklearn.preprocessing import Normalizer

Normalizer().fit_transform(X)
```

### 数据集划分

#### 留出法

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,y,train_size=0.8)
train_x.shape
```

#### k折交叉验证

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=10)
for train,test in kf.split(X,y):
    print(train,test)#train,test 指的是索引
```

### 模型使用

```python
#分类的模型：
# from sklearn.tree import DecisionTreeClassifier#决策树分类
from sklearn.ensemble import RandomForestClassifier  #随机森林分类
# from sklearn.linear_model import LogisticRegressionCV #逻辑回归分类
# from sklearn.svm import SVC #支持向量机分类
# from sklearn.neural_network import MLPClassifier  #多层感知机分类
RFmodel=RandomForestClassifier()
RFmodel.fit(train_x,train_y)
pre_y=RFmodel.predict(test_x)  #返回预测的分类
pre_proba_y=RFmodel.predict_proba(test_x)  #返回预测的可能性
pre_y
```

```python
pre_proba_y
```

### 模型评价

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(test_y,pre_y)
```

```python
confusion_matrix(test_y,pre_y) #混淆矩阵
```

```python
print(classification_report(test_y,pre_y))  #分类性能报告
```

## K折交叉验证

```python
#导入数据
from sklearn.datasets import load_diabetes
X,Y=load_diabetes()['data'],load_diabetes()['target']
#预处理
#降维（选做）
from sklearn.decomposition import PCA
X_new=PCA(n_components=5).fit_transform(X)
#标准化（选做）
from sklearn.preprocessing import Normalizer
X_PN=Normalizer().fit_transform(X_new)

#数据集划分
from sklearn.model_selection import KFold
kf=KFold(n_splits=10)
#导入模型并训练
from sklearn.tree import DecisionTreeRegressor
DTmodel=DecisionTreeRegressor()
#模型评价
from sklearn.metrics import mean_absolute_error
for train,test in kf.split(X_PN,Y):
    DTmodel.fit(X_PN[train,:],Y[train])
    pre_y=DTmodel.predict(X_PN[test,:])
    print(mean_absolute_error(Y[test],pre_y))
```

## 留出法

```python
#导入数据
from sklearn.datasets import load_diabetes
X,Y=load_diabetes()['data'],load_diabetes()['target']
#预处理
#降维（选做）
from sklearn.decomposition import PCA
X_new=PCA(n_components=5).fit_transform(X)
#标准化（选做）
from sklearn.preprocessing import Normalizer
X_PN=Normalizer().fit_transform(X_new)

#数据集划分
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X_PN,Y,test_size=0.8)
#导入模型并训练
from sklearn.tree import DecisionTreeRegressor
DTmodel=DecisionTreeRegressor()
DTmodel.fit(train_x,train_y)
pre_y=DTmodel.predict(test_x)
#模型评价
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(test_y,pre_y))
```
