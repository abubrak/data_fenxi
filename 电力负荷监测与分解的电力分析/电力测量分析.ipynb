```python
def fun(type_,s_n,UI):
    #提取第几个电器数据
    index_=all_data[s_n]['type']==type_
    #周波/谐波
    type_data=all_data[s_n][index_]
    #提取电流电压
    col=[i for i in type_data.columns if UI in i]
    UIC_data=type_data[col]
    #绘制动图
    fig=plt.figure()
    ims=[]
    for i in UIC_data.index:
        plt.xticks([0,len(col)],[col[0],col[-1]])
        im=plt.plot(UIC_data.loc[i,:])
        plt.title(f'{type_}{UI}{s_n}变化')
        ims.append(im)
    ani=animation.ArtistAnimation(fig,ims,interval=200,repeat_delay=1000)
    ani.save(f'{type_}{UI}{s_n}变化图.gif',writer='pillow')
```

### 识别数据

### 周波数据

```python
# 一个数据为例，做验证
zhoubo=all_data['周波数据']
cdata=zhoubo.iloc[0,1:129].values  #提取电流数据
#为了带入模型，需要对数据维度改变
# cdata.shape #(128,)
cdata=cdata.reshape(128,1)
cdata.shape
```

```python
from sklearn.cluster import KMeans
kmodel=KMeans(n_clusters=2)
kmodel.fit(cdata)
```

```python
kmodel.cluster_centers_
```

```python
print((kmodel.labels_==1).sum())
print((kmodel.labels_==0).sum())
```

```python
def fun(x):
    x=x.values
    x=x.reshape(128,1)
    kmodel=KMeans(n_clusters=2)
    kmodel.fit(x)
    cluster_data=kmodel.cluster_centers_
    low=cluster_data.min() #波谷
    up=cluster_data.max() #波峰
    #获取波峰波谷对应的样本数
    low_num=(kmodel.labels_==cluster_data.argmin()).sum()
    up_num=(kmodel.labels_==cluster_data.argmax()).sum()
    return low,up,low_num,up_num
```

```python
# 获取所有的IC数据
zhoubo=all_data['周波数据']
ans=zhoubo.iloc[:,1:129].apply(fun,axis=1)#axis=1按例进行
```

```python
zhoubo['波谷'],zhoubo['波峰'],zhoubo['波谷_num'],zhoubo['波峰_num']=zip(*ans)
zhoubo
```

### 谐波数据

```python
xiebp=all_data['谐波数据']
col_lis=[i for i in xiebp.columns if 'IC' in i]
tmp=xiebp[col_lis]
xiebp_ans=tmp.apply(lambda d:
                   [d.max(),d.min(),d.std(),
                   d.kurt(),#峰度
                   d.skew(),#偏度
                   d.median(),
                   d.quantile(0.9)],axis=1)
```

```python
xiebp_ans
```

```python
xiebp_ans=pd.DataFrame(
    list(xiebp_ans.values),
        index=xiebp_ans.index,
        columns=['最大值','均值','标准差','峰度','偏度','中位数','0.9分位数'])
```

```python
xiebp_ans
```
